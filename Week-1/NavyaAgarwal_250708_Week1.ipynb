Write a function defintion that:

```python
def concat(a,b):
  c=a+b
  return c
print(concat("hello","!"))
```

```python
def SumOfSquares(a,b):
  c=a*a+b*b
  return c
print(SumOfSquares(2,3))
```

```python
def CountVowel(a):
  vowels="aeiouAEIOU"
  vc=0
  for ch in a:
    if ch in vowels:
      vc+=1
  return vc
print(CountVowel("hello"))

```

```python
def WithoutVowel(a):
  vowels="aeiouAEIOU"
  newstr=""
  for ch in a:
    if ch not in vowels:
      newstr+=ch
  return newstr
print(WithoutVowel("hello"))
```

```python
def ShortestString(s):
  shortest=len(s[0])
  for i in s:
    if len(i)<shortest:
      shortest=len(i)
  return shortest
print(ShortestString(["hello","hey","hi"]))
```

```python
def oddlist(l):
  ll=[]
  for i in l:
    if i%2!=0:
      ll.append(i)
  return ll
print(oddlist([1,2,3,4,5]))

```

```python
def AddList(l):
  sum=0
  for i in l:
    sum+=i
  return sum
print(AddList([1,2,3,4]))
```

**Perform the following operations on the csv file:**

1. Read the csv into a pandas dataframe and back it up (you will thank yourself for doing this when you mess up something while manipulating the dataframe).
2. Explore the dataframe (find out the different ways you can do this).
3. Play around with each column of the dataframe and figure out ways to modify them according to your needs.
4. Read up on ``iloc`` and demonstrate it on the dataframe.
5. Handle ``null`` values in the dataframe.
6. Try dropping rows/columns.
7. Filter rows where a string column starts with a specific letter.
8. Create a new column as a combination of two existing columns (this is feature engineering which you may use in your ML journey).

```python
csv_text = """id,first_name,last_name,age,salary,city,department
1,navya,agarwal,19,50000,Bareilly,HR
2,moksh,bhatt,18,62000,Bharuch,Engineering
3,tamanna,gupta,,58000,Noida,Marketing
4,anshika,agarwal,19,,Lucknow,Finance
5,rishu,prakas,23,54000,bihar,Engineering
6,nitin,agarwal,43,30000,Chandausi,Intern
7,sugandh,goel,40,70000,gurgoan,HR
8,uma,goel,64,68000,surat,Operations
9,surya,prakash,,47000,kanpur,Marketing
10,swati,chaudary,37,60000,delhi,Finance
"""

with open("employees.csv", "w") as f:
    f.write(csv_text)

import pandas as pd

# 1. Read the CSV into a pandas dataframe
df = pd.read_csv("employees.csv")
df_backup = df.copy()

print("=== Original DataFrame ===")
print(df)
print("\n")

# 2. Explore the dataframe in different ways

print("First 5 rows (head)")
print(df.head())
print("\n")

print(" Last 5 rows (tail) ")
print(df.tail())
print("\n")

print(" Shape of the DataFrame (rows, columns) ")
print(df.shape)
print("\n")

print(" Column names ")
print(df.columns)
print("\n")

print(" Data types of each column ")
print(df.dtypes)
print("\n")

print(" Info about DataFrame ")
print(df.info())
print("\n")

print(" Summary statistics for numeric columns ")
print(df.describe())
print("\n")

# 3. Play around with each column

print(" Age column")
print(df["age"])
print("\n")

df["age_plus_5"] = df["age"] + 5
print("New column: age_plus_5 ")
print(df[["age", "age_plus_5"]])
print("\n")

print("Salary column ")
print(df["salary"])
print("\n")

df["salary_k"] = df["salary"] / 1000
print(" New column: salary_k ")
print(df[["salary", "salary_k"]])
print("\n")

print(" First names ")
print(df["first_name"])
print("\n")

df["first_name_upper"] = df["first_name"].str.upper()
print(" New column: first_name_upper ")
print(df[["first_name", "first_name_upper"]])
print("\n")

df["city_length"] = df["city"].str.len()
print("New column: city_length ")
print(df[["city", "city_length"]])
print("\n")

# 4. Using iloc (position-based indexing)

print("Using iloc examples ")

print("First row (df.iloc[0]):")
print(df.iloc[0])
print("\n")

print("First 3 rows (df.iloc[0:3]):")
print(df.iloc[0:3])
print("\n")

print("First row, first column (df.iloc[0, 0]):")
print(df.iloc[0, 0])
print("\n")

print("First 5 rows, first 2 columns (df.iloc[0:5, 0:2]):")
print(df.iloc[0:5, 0:2])
print("\n")

print("All rows, second column (df.iloc[:, 1]):")
print(df.iloc[:, 1])
print("\n")

print("Last row (df.iloc[-1]):")
print(df.iloc[-1])
print("\n")

# 5. Handle null values

print("=== Check for null values in each column ===")
print(df.isna().sum())
print("\n")

mean_age = df["age"].mean()
print("Mean age (ignoring NaN):", mean_age)
df["age_filled"] = df["age"].fillna(mean_age)
print(" Age and age_filled columns ")
print(df[["age", "age_filled"]])
print("\n")

df["salary_filled"] = df["salary"].fillna(0)
print(" Salary and salary_filled columns ")
print(df[["salary", "salary_filled"]])
print("\n")

# 6. Try dropping rows/columns

df_dropped_column = df.drop(columns=["city_length"])
print(" DataFrame after dropping 'city_length' column")
print(df_dropped_column.head())
print("\n")

df_dropped_row = df.drop(index=0)
print(" DataFrame after dropping row with index 0 ")
print(df_dropped_row.head())
print("\n")

# 7. Filter rows where a string column starts with a specific letter

filtered_A = df[df["first_name"].str.startswith("A")]
print("=== Rows where first_name starts with 'A' ===")
print(filtered_A)
print("\n")

filtered_J = df[df["first_name"].str.startswith("J")]
print("=== Rows where first_name starts with 'J' ===")
print(filtered_J)
print("\n")

# 8. Create a new column as a combination of two existing columns

df["full_name"] = df["first_name"] + " " + df["last_name"]
print("=== New column: full_name ===")
print(df[["first_name", "last_name", "full_name"]])
print("\n")

df["city_department"] = df["city"] + " - " + df["department"]
print("=== New column: city_department ===")
print(df[["city", "department", "city_department"]])
print("\n")

print("=== Final DataFrame with new columns ===")
print(df.head())
```
