```python
# Original list operations
print(":)Using built-in methods")
l = [1, 12, 43, 4, 15, 26]
print("1. Initial list:", l)

# 2. Append a no
l.append(100)
print("2. List after appending 100:", l)

# 3. Sort
l.sort()
print("3. List after sorting:", l)

# 4. Reverse the sorted list
l.reverse()
print("4. List after reversing sorted list:", l)

print("\n:)Repeating with loops and list indexing")
# 1. this list
l = [1, 12, 43, 4, 15, 26]
print("1. Initial list (for loop operations):", l)

# 2. Append a number
l.append(100)
print("2. List after appending 100 (loop operations):", l)

# 3. Sort (Bubble Sort)
n = len(l)
for i in range(n):
    for j in range(0, n - i - 1):
        if l[j] > l[j + 1]:
            l[j], l[j + 1] = l[j + 1], l[j]
print("3. List after bubble sort:", l)

# 4. Reverse the sorted list
left = 0
right = len(l) - 1
while (left < right):
    l[left], l[right] = l[right], l[left]
    left += 1
    right -= 1
print("4. List after reversing (indexing):",l)
```

***Functions***

(Tip: Before starting this section, brush up on strings in python)

Write a function defintion that:
1. Adds the string "Hello, " to the beginning and "!" to the end of any given input.
2. Takes in two numbers, squares each number, then returns the sum of both squares.
3. Takes in value and returns the count of the nubmer of vowels in a sequence.
4. Takes in string and returns the string without any vowels.
5. Takes in a list of strings and returns the shortest string in the list.
6. Takes in sequence of numbers and returns the odd numbers in a list.
7. Takes in sequence of numbers and returns all the numbers added together.

```python
# 1. Adds the string "Hello, " to the beginning and "!" to the end of any given input.
def add_greeting(input_string):
    return "Hello, " + str(input_string) + "!"

# 2. Takes in two numbers, squares each number, then returns the sum of both squares.
def sum_of_squares(num1, num2):
    return num1**2 + num2**2

# 3. Takes in value and returns the count of the number of vowels in a sequence.
def count_vowels(sequence):
    vowels = "aeiouAEIOU"
    count = 0
    for char in str(sequence):
        if char in vowels:
            count += 1
    return count

# 4. Takes in string and returns the string without any vowels.
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join([char for char in input_string if char not in vowels])

# 5. Takes in a list of strings and returns the shortest string in the list.
def shortest_string(list_of_strings):
    if not list_of_strings:
        return None # or raise an error for an empty list
    return min(list_of_strings, key=len)

# 6. Takes in sequence of numbers and returns the odd numbers in a list.
def get_odd_numbers(sequence_of_numbers):
    return [num for num in sequence_of_numbers if num % 2 != 0]

# 7. Takes in sequence of numbers and returns all the numbers added together.
def sum_sequence(sequence_of_numbers):
    return sum(sequence_of_numbers)
```

***Sets***

(Read up on sets in python)

Write a function definition that:
1. Takes in a list and returns a set with only the unique values from that list.
2. Takes two lists and returns a single set with the values that each list has in common.

(Bonus - Try doing the above questions from scratch using loops)

```python
# 1. Takes in a list and returns a set with only the unique values from that list.
def get_unique_values(input_list):
    return set(input_list)

# 2. Takes two lists and returns a single set with the values that each list has in common.
def get_common_values(list1, list2):
    return set(list1).intersection(set(list2))

print("--- Using built-in set methods ---")
my_list = [1, 2, 2, 3, 4, 4, 5]
print(f"Unique values from {my_list}: {get_unique_values(my_list)}")

list_a = [1, 2, 3, 4, 5]
list_b = [4, 5, 6, 7, 8]
print(f"Common values between {list_a} and {list_b}: {get_common_values(list_a, list_b)}")


# Bonus - Try doing the above questions from scratch using loops
print("\n--- Bonus: Using loops ---")

def get_unique_values_loop(input_list):
    unique_elements = []
    for item in input_list:
        if item not in unique_elements:
            unique_elements.append(item)
    return set(unique_elements)

def get_common_values_loop(list1, list2):
    common_elements = []
    for item1 in list1:
        for item2 in list2:
            if item1 == item2 and item1 not in common_elements:
                common_elements.append(item1)
    return set(common_elements)

# Example usage for bonus functions:
print(f"Unique values from {my_list} (loop): {get_unique_values_loop(my_list)}")
print(f"Common values between {list_a} and {list_b} (loop): {get_common_values_loop(list_a, list_b)}")
```
